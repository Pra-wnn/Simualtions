```python
# Simualte Gmabler's Ruin

import numpy as np 
import matplotlib.pyplot as plt
import random 
import pandas as pd 
```

```python
def gamblers_ruin():
    gambling_money = 50;
    gamblin_goal = 100;
    gambling_simulation = [];

    while gambling_money > 0 and gambling_money < gamblin_goal:
        if random.randint(0,1) == 0:
            gambling_money += 1
        else:
            gambling_money -= 1
        gambling_simulation.append(gambling_money)
    return gambling_simulation
```

```python
plt.plot(gamblers_ruin())
plt.yticks(np.arange(-20,120,10))
plt.axhline(y=0, color='r', linestyle='-')
plt.axhline(y=100, color='black', linestyle='-')
plt.xlabel('Number of bets')
plt.ylabel('Winnings')
plt.title('Gambling Simulation');
```
